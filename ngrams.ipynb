```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

import nltk
nltk.download('punkt')
nltk.download('stopwords')

from nltk.stem import SnowballStemmer
snowball = SnowballStemmer(language="russian")
russian_stopwords = stopwords.words("russian") + ["это", "[", "]", "спасибо", "который", "всё", "все"]
```

```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
import nltk
from nltk.util import ngrams
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures
```

```python
def normalized_transcript(report):
    if report['video_transcript'] is None: return
    r_words_spoken = word_tokenize(report['video_transcript'], language="russian")
    
    text = ""
    for w in r_words_spoken:
        w = morph.parse(w)[0].normal_form
        if w in russian_stopwords: continue
        text += (w + " ")

    return text
```

```python
df['normalized_transcript'] = df.apply(normalized_transcript, axis=1)
```

```python
df
```

```python
def bigrams(report):
    if report['normalized_transcript'] is None: return
    text = report['normalized_transcript'].split()
    bigrams = nltk.bigrams(text)
    bigram_fd = nltk.FreqDist(bigrams)

    return bigram_fd.most_common()
```

```python
df['bigrams'] = df.apply(bigrams, axis=1)
```

```python
words_spoken = {}
def yearly_grams(report):
    if report['bigrams'] is None: return
    year = report['year']
    if year not in words_spoken: words_spoken[year] = {}
    r_words_spoken = report['bigrams']
    for (r_w, r_c) in r_words_spoken:
        if r_w not in words_spoken[year]: words_spoken[year][r_w] = 0
        words_spoken[year][r_w] += r_c

df.apply(yearly_grams, axis=1)
```
